```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from powerbiclient import Report
```

```python
df1 = pd.read_csv("/population.csv")
```

```python
df2 = pd.read_csv("/households.csv")
```

```python
population = df1.iloc[-111:]
```

```python
households= df2.iloc[-111:]
```

```python
households.head()
```

```python
data = population.Pop
labels = population.Ward_ID

plt.figure(figsize=(100,15))

plt.xticks(range(len(data)), labels)
plt.xlabel('Ward_ID')
plt.ylabel('Population Size')
plt.title('Population Size Per Ward')
plt.bar(range(len(data)), data) 
plt.show()
```

##Checking the population distribution in each of the 111 Wards of Capetown

```python
population.plot(x="Ward_ID", y="Pop", kind="bar", figsize=(50, 15))
```

##Tenure Status

```python
#data = [households['Rented'].sum(),households['Owned_but_'].sum(),households['Occupied_r'].sum(),households['Owned_and_'].sum(),households['Other1'].sum()]
data = households[['Rented','Owned_but_','Occupied_r','Owned_and_','Other1']].sum()
my_labels = "Rented","Owned but not paid off","Occupied Rent Free","Owned and fully paid","Other"

plt.figure(figsize=(12,12))

plt.pie(data,labels=my_labels,autopct='%1.1f%%')
plt.title('Tenure Status')
plt.axis('equal')
plt.show()
```

##Household Size

```python
data2 = households[['HH_Size_1','HH_Size_2','HH_Size_3','HH_Size_4','HH_Size_5','HH_Size_6','HH_Size_7','HH_Size_8','HH_Size_9','HH_Size_10']].sum()
my_labels2 = 'HH_Size_1','HH_Size_2','HH_Size_3','HH_Size_4','HH_Size_5','HH_Size_6','HH_Size_7','HH_Size_8','HH_Size_9','HH_Size_10'
y_pos = np.arange(len(my_labels2))

plt.figure(figsize=(20,10))

plt.bar(y_pos, data2, align='center', alpha=0.5)
plt.xticks(y_pos, my_labels2)
plt.ylabel('Number of Households')
plt.title('Household size')

plt.show()
```

##Average Household Income

```python
data3 = households[['AHHI_No_in','AHHI_R_1__','AHHI_R_480','AHHI_R_960','AHHI_R_19_','AHHI_R_38_','AHHI_R_76_','AHHI_R_153','AHHI_R_307','AHHI_R_614','AHHI_R_1_2','AHHI_R_2_4','AHHI_Unspe']].sum()
my_labels3 = 'No income','R1-R4,800','R4,801-R9,600','R9,601-R19,600','R19,601-R38,200','R38,201-R76,400','R76,401-R153,800','R153,801-R307,600','R307,601-R614,400','R614,401-1,228,800','R1,228,801-R2,457,600','R2,457,601+','Unspecified'
y_pos1 = np.arange(len(my_labels3))

plt.figure(figsize=(30,10))

plt.bar(y_pos1, data3, align='center', alpha=0.5)
plt.xticks(y_pos1, my_labels3)
plt.ylabel('Number of Households')
plt.title('Average Household Income')

plt.show()
```

##Employment Status

```python
data4 = population[['Employed','Unemployed','Discourage','Other_NotE']].sum()
my_labels4 = "Employed", "Unemployed", "Discouraged Work Seeker", "Not Economically Active"

plt.figure(figsize=(15,11))

plt.pie(data4,labels=my_labels4,autopct='%1.1f%%')
plt.title('Employment Status')
plt.axis('equal')
plt.show()
```

##Personal Monthly Income

```python
data5 = population[['PMI_No_inc','PMI_R_1___','PMI_R_401_','PMI_R_801_','PMI_R_1_60','PMI_R_3_20','PMI_R_6_40','PMI_R_12_8','PMI_R_25_6','PMI_R51201','PMI_R10240','PMI_R20480','PMI_Unspec']].sum()
my_labels5 = 'PMI_No_inc','PMI_R_1___','PMI_R_401_','PMI_R_801_','PMI_R_1_60','PMI_R_3_20','PMI_R_6_40','PMI_R_12_8','PMI_R_25_6','PMI_R51201','PMI_R10240','PMI_R20480','Unspecified'
y_pos2 = np.arange(len(my_labels5))

plt.figure(figsize=(30,10))

plt.bar(y_pos2, data5, align='center', alpha=0.5)
plt.xticks(y_pos2, my_labels5)
plt.ylabel('Number of People')
plt.title('Personal Monthly Income')

plt.show()
```
